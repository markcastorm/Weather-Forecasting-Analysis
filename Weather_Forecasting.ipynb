```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

data = pd.read_csv("DailyDelhiClimateTrain.csv")
print(data.head)
```

Now lets see the mean temperature in Delhi over the years:

```python
figure = px.line(data, x="date",
                 y="meantemp",
                 title="Mean Temprature in Delhi Over the Years")
figure.show()
```

Now let’s see the humidity in Delhi over the years

```python
figure = px.line(data, x="date",
                 y="humidity",
                 title="Humidity in Delhi Over the year")
figure.show()
```

Now let’s check the wind speed in Delhi over the years:

```python
figure = px.line(data, x="date",
                 y="wind_speed",
                 title='Wind Speed in Delhi Over the Years ')
figure.show()
```

From the results we can see that till 2015 the wind speed was higher during (August & september) and it retreats during (December & January)

Now let’s check on the relationship between temperature and humidity:

```python
figure = px.scatter(data_frame = data, x="humidity",
                    y="meantemp", size="meantemp",
                    trendline="ols",
                    title ="Relationship Between Temprature and humidity")
figure.show()
```

From the visualze above we can see that there’s a negative correlation between temperature and humidity ,the higher the tempreture the lower the humidity

Now let’s analyze the temperature change


First lets change the datatype from the date column

```python
data["date"] = pd.to_datetime(data["date"], format = '%Y-%m-%d')
data['year'] = data['date'].dt.year
data["month"] = data["date"].dt.month
print(data.head())
```

Now lets plot the temprature change over the years

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(15, 10))
plt.title("Temperature Change in Delhi Over the Years")
sns.lineplot(data = data, x='month', y='meantemp', hue='year')
plt.show()
```

Lets convert the data into a format that the prophet model can accept as "ds" and label as "y"

```python
forecast_data = data.rename(columns = {"date": "ds",
                                       "meantemp": "y"})
print(forecast_data)
```

Now lets use the prophet for weather forecasting

```python
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
model = Prophet()
model.fit(forecast_data)
forecasts = model.make_future_dataframe(periods=365)
predictions = model.predict(forecasts)
plot_plotly(model, predictions)
```
